## Visualizing Stock Trends

```python
google = pd.read_csv('GOOGL_data.csv')
apple = pd.read_csv('AAPL_data.csv')
facebook = pd.read_csv('FB_data.csv')
amazon = pd.read_csv('AMZN_data.csv')
microsoft = pd.read_csv('MSFT_data.csv')
```

```python
google.head()
```

```python
google.shape
```

```python
plt.figure(figsize=(16, 8), dpi=300)

plt.plot('date', 'close', data=google, label='Google')
plt.plot('date', 'close', data=apple, label='Apple')
plt.plot('date', 'close', data=facebook, label='Facebook')
plt.plot('date', 'close', data=amazon, label='Amazon')
plt.plot('date', 'close', data=microsoft, label='Microsoft')

plt.xticks(ticks=np.arange(0, 1260, step=40), rotation=70)
plt.yticks(ticks=np.arange(0, 1450, step=100))

plt.title('Stock trend', fontsize=16)

plt.ylabel('Closing price in $', fontsize=14)

plt.legend()  # show legend based on labels in plt.plot

plt.grid()  # show grid

plt.show()  # show plot

plt.close()
```

## Bar Plot
#### plt.bar(x, height, [width]), plt.barh()

```python
movies = pd.read_csv('movie_scores.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
plt.figure(figsize=(10, 5), dpi=300)

labels = movies['MovieTitle']
x = np.arange(len(labels))

width = 0.4

plt.bar(x - width / 2, movies['Tomatometer'], width=width, label='Tomatometer')
plt.bar(x + width / 2, movies['AudienceScore'], width=width, label='Audience Score')

plt.xticks(x, rotation=15)
plt.yticks(np.arange(0, 101, 20))

ax = plt.gca()

ax.set_xticklabels(labels)
ax.set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])

ax.set_yticks(np.arange(0, 100, 5), minor=True)

ax.yaxis.grid(which='major')
ax.yaxis.grid(which='minor', linestyle='--')

plt.title('Movie comparison')

plt.legend()

plt.show()
```

## Stacked Bar Chart
#### plt.bar(x, bars1), plt.bar(x, bars2, bottom=bars1), plt.bar(x, bars3, bottom=np.add(bars1, bars2))

```python
import seaborn as sns
```

```python
bills = sns.load_dataset('tips')
```

```python
bills
```

```python
days = bills['day'].unique()
days
```

```python
days_range = np.arange(len(days))
days_range
```

```python
smoker = bills['smoker'].unique()
smoker
```

```python
bills_by_days = [bills[bills['day'] == day] for day in days]
bills_by_days
```

```python
bills_by_smoker = [bills[bills['smoker'] == yesno] for yesno in smoker]
bills_by_smoker
```

```python
bills_by_days_smoker = [[bills_by_days[day][bills_by_days[day]['smoker'] == yesno] for yesno in smoker] for day in days_range]
bills_by_days_smoker
```

```python
total_by_days_smoker = [[bills_by_days_smoker[day][yesno]['total_bill'].sum() for yesno in range(len(smoker))] for day in days_range]
total_by_days_smoker
```

```python
totals = np.asarray(total_by_days_smoker)
totals
```

```python
plt.figure(figsize=(10, 5), dpi=300)

plt.bar(days_range, totals[:, 0], label='Non-Smoker')
plt.bar(days_range, totals[:, 1], bottom=totals[:, 0], label='Smoker')

plt.legend()

plt.xticks(days_range)
ax = plt.gca()
ax.set_xticklabels(days)
ax.yaxis.grid()
plt.ylabel('Daily total sales in $')
plt.title('Restaurant performance')

plt.show()
```

## Scatter Plot
#### plt.scatter(x, y)

```python
data = pd.read_csv('anage_data.csv')
```

```python
data.head()
```

```python
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
data = data[np.isfinite(data[longevity]) & np.isfinite(data[mass])]
```

```python
amphibia = data[data['Class'] == 'Amphibia']
aves = data[data['Class'] == 'Aves']
mammalia = data[data['Class'] == 'Mammalia']
reptilia = data[data['Class'] == 'Reptilia']
```

```python
plt.figure(figsize=(10,6), dpi=300)

plt.scatter(amphibia[mass], amphibia[longevity], label='Amphibia')
plt.scatter(aves[mass], aves[longevity], label='Aves')
plt.scatter(mammalia[mass], mammalia[longevity], label='Mammalia')
plt.scatter(reptilia[mass], reptilia[longevity], label='Reptilia')

ax = plt.gca()

ax.set_xscale('log')    
ax.set_yscale('log') 

plt.xlabel('Body mass in grams')
plt.ylabel('Maximum longevity in years')

plt.legend()

plt.show()
```

## Radar Charts

```python
data2 = pd.DataFrame({
    'Employee': ['Alex', 'Alice', 'Chris', 'Jennifer'],
    'Efficiency': [5, 4, 4, 3,],
    'Quality': [5, 5, 3, 3],
    'Commitment': [5, 4, 4, 4],
    'Responsible Conduct': [4, 4, 4, 3],
    'Cooperation': [4, 3, 4, 5]
})
```

```python
attributes = list(data2.columns[1:])
values = list(data2.values[:, 1:])
employees = list(data2.values[:, 0])

angles = [n / float(len(attributes)) * 2 * np.pi for n in range(len(attributes))]

angles += angles[:1]
values = np.asarray(values)
values = np.concatenate([values, values[:, 0:1]], axis=1)
```

```python
plt.figure(figsize=(8, 8), dpi=150)

for i in range(4):
    ax = plt.subplot(2, 2, i + 1, polar=True)
    ax.plot(angles, values[i])
    ax.set_yticks([1, 2, 3, 4, 5])
    ax.set_xticks(angles)
    ax.set_xticklabels(['Quality', 'Quality', 'Commitment', 'Responsible Conduct', 'Cooperation', 'Efficiency'])
    ax.set_title(employees[i], fontsize=14, color='r')  
    
plt.tight_layout()

plt.show()
```

## Scatter Plot with Marginal Histograms

```python
data3 = pd.read_csv('anage_data.csv')
```

```python
data3.shape
```

```python
data3
```

```python
# Preprocessing
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
data = data3[np.isfinite(data3[longevity]) & np.isfinite(data3[mass])]
```

```python
# Sort according to class
data3 = data3[data3['Class'] == 'Aves']
data3 = data3[data3[mass] < 20000]
```

```python
data3
```

```python
# Create figure
fig = plt.figure(figsize=(8, 8), dpi=150)
# Create gridspec
gs = fig.add_gridspec(4, 4)
# Specify subplots
histx_ax = fig.add_subplot(gs[0, 0:3])
histy_ax = fig.add_subplot(gs[1:, 3])
scatter_ax = fig.add_subplot(gs[1:, 0:3])
# Create plots
scatter_ax.scatter(data3[mass], data3[longevity])
histx_ax.hist(data3[mass], bins=20, density=True)
histx_ax.set_xticks([])
histy_ax.hist(data3[longevity], bins=20, density=True, orientation='horizontal')
histy_ax.set_yticks([])
# Add labels and title
plt.xlabel('Body mass in grams')
plt.ylabel('Maximum longevity in years')
fig.suptitle('Scatter plot with marginal histograms')
# Show plot
plt.tight_layout()
plt.show()
```
