### iterating
#### np.nditer(...), np.ndenumerate(...)

```python
dataset2 = np.genfromtxt('normal_distribution_splittable.csv', delimiter=',')
```

```python
dataset2
```

```python
second_row = dataset2[1]
np.mean(second_row)
```

```python
last_row = dataset2[-1]
np.mean(last_row)
```

```python
first_val_first_row = dataset2[0][0]
np.mean(first_val_first_row)
```

```python
every_other_elem = dataset2[4, ::2]
np.mean(every_other_elem)
```

```python
index = 0
for x in np.nditer(dataset2):
    print(x, index)
    index += 1
```

```python
for index, value in np.ndenumerate(dataset2):
    print(index, value)
```

### Filtering
#### dataset[dataset > ...], np.extraxt((dataset > ...) & (dataset < ...)), np.where

```python
values_greater_105 = dataset[dataset > 105]
values_greater_105
```

```python
values_between_90_95 = np.extract((dataset > 90) & (dataset < 95), dataset)
values_between_90_95
```

```python
rows, cols = np.where(abs(dataset - 100) < 1)
one_away_index = [[rows[index], cols[index]] for (index, _) in np.ndenumerate(rows)]
one_away_index
```

### Sorting
#### np.sort, np.argsort

```python
row_sorted = np.sort(dataset)  # last axis by default, i.e. axis=1 (rows)
row_sorted
```

```python
col_sorted = np.sort(dataset, axis=0)
col_sorted
```

```python
index_sorted = np.argsort(dataset[0])
dataset[0][index_sorted]
```
