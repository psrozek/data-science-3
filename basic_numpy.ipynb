### iterating
#### np.nditer(...), np.ndenumerate(...)

```python
dataset2 = np.genfromtxt('../Python/data_visualization/normal_distribution_splittable.csv', delimiter=',')
```

```python
dataset2
```

```python
second_row = dataset2[1]
np.mean(second_row)
```

```python
last_row = dataset2[-1]
np.mean(last_row)
```

```python
first_val_first_row = dataset2[0][0]
np.mean(first_val_first_row)
```

```python
every_other_elem = dataset2[4, ::2]
np.mean(every_other_elem)
```

```python
index = 0
for x in np.nditer(dataset2):
    print(x, index)
    index += 1
```

```python
for index, value in np.ndenumerate(dataset2):
    print(index, value)
```
