```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
data = pd.DataFrame({'Groups' : ['Group A'] * len(group_a) + ['Group B'] * len(group_b) + ['Group C'] \
                     * len(group_c) + ['Group D'] * len(group_d), 'IQ score' : group_a + group_b + group_c + group_d})
```

```python
data
```

```python
plt.figure(dpi=150)

sns.set_style('whitegrid')

sns.boxplot(x='Groups', y='IQ score', data=data)

sns.despine(left=True, right=True, top=True)

plt.title('IQ scores for different test groups')

plt.show()
```

## FacetGrid
#### sns.FacetGrid(data, row, col, hue, …), FacetGrid.map(func, *args, **kwargs)

```python
yt = pd.read_csv('YouTube.csv')
```

```python
yt.head()
```

```python
subs = yt[yt.columns[1]].tolist()
views = yt[yt.columns[2]].tolist()
channel = yt[yt.columns[0]].tolist()
```

```python
yt = pd.DataFrame({'YouTube Channel' : channel + channel, \
                   'Subs/Views (in millions)' : subs + views, 'Type' : ['Subscribers'] * len(subs) + ['Viewers'] * len(views)})
```

```python
yt
```

```python
sns.set()
g = sns.FacetGrid(yt, col='Type', hue='Type', sharex=False, height=8)
g.map(sns.barplot, 'Subs/Views (in millions)', 'YouTube Channel')
plt.show()
```

## Regression Plots
#### sns.regplot()

```python
anage = pd.read_csv('anage_data.csv')
```

```python
anage
```

```python
anage = anage[anage['Class'] == 'Mammalia']
```

```python
anage = anage[anage['Body mass (g)'] < 200000]
```

```python
anage = anage[np.isfinite(anage['Maximum longevity (yrs)']) & np.isfinite(anage['Body mass (g)'])]
```

```python
x = anage['Body mass (g)']
y = anage['Maximum longevity (yrs)']
```

```python
sns.set()
plt.figure(figsize=(10, 6), dpi=300)

sns.regplot(x=x, y=y)

plt.show()
```

## Squarify

```python
import squarify
```

```python
salary = pd.read_csv('age_salary_hours.csv')
```

```python
salary
```

```python
degrees = set(salary['Education'])
percentages = []
for degree in degrees:
    percentages.append(salary[salary['Education'] == degree].shape[0])
percentages = np.array(percentages)
percentages = ((percentages / percentages.sum()) *100)
```

```python
labels = [degree + ' ({0:.1f}%)'.format(percentage) for degree, percentage in zip(degrees, percentages)]
```

```python
sns.set()
plt.figure(figsize=(9, 6), dpi=200)

squarify.plot(percentages, label=labels, color=sns.color_palette('colorblind', len(degrees)))

plt.axis('off')

plt.title('Degrees')

plt.show()
```
