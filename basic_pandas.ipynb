## Analyzing forest fires dataset

```python
forest_fire = pd.read_csv('forestfires.csv')  # read file
```

```python
forest_fire.head(3)  # show first three rows
```

```python
forest_fire.shape
```

```python
forest_fire = forest_fire[(forest_fire["area"] > 0)]  # exclude fires with area = 0
```

```python
forest_fire
```

```python
forest_fire["area"].mean()  # average area
```

```python
forest_fire["area"].min()  # min. area
```

```python
forest_fire["area"].max()  # max. area
```

```python
forest_fire["area"].std()  # area standard deviation
```

```python
forest_fire.sort_values(by=["area"]).tail(20)  # sort values by area and show maximum 20
```

```python
forest_fire["area"].median()  # area median
```

```python
months = forest_fire["month"].unique()  # list of unique months
months
```

```python
forest_fire[forest_fire["month"] == "mar"].shape[0]  # fire cases in March
```

```python
for month in months:  # calculate fire cases and average temperature for each unique month
    month_forest_fire = forest_fire[forest_fire["month"] == month]
    fires_in_month = month_forest_fire.shape[0]
    avg_temp_in_month = int(month_forest_fire["temp"].mean())
    
    print(str(fires_in_month) + " fires in " + month + " with a mean temperature of ~" + str(avg_temp_in_month) + " centigrades")
```
